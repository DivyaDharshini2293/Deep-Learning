# **Terrain recognition**

```python
classes = ["sandy","rocky","grass","marshy"]
classes = ["sandy","rocky","grass","marshy"]
url = "https://www.kongu.ac.in/webalbum/img/mba.jpg"
url1 = "https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcTymKlkPWsIigRmQEgxRaY8ZvTZxcK_AWsbOuLZLW-UoQC2UI60Ae3VkRQmM3Gvn9CkoYg&usqp=CAU"
```

```python
#Do not change this
import io
from PIL import Image
import requests
from transformers import CLIPProcessor, CLIPModel
model = CLIPModel.from_pretrained("openai/clip-vit-large-patch14")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-large-patch14")
```

```python
image = Image.open(requests.get(url, stream=True).raw)
inputs = processor(text=classes, images=image, return_tensors="pt", padding=True)
probs = model(**inputs).logits_per_image.softmax(dim=1)
print("Class = " + classes[((probs.tolist())[0]).index(max((probs.tolist())[0]))])
image
```
